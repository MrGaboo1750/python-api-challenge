```python
# WeatherPy
# Written by Jason Gabunilas

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from time import strftime, localtime
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
# for city in cities:
#     print(city)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# initiate empty lists for city data
city_ID = []
cities_added = []
latitude = []
longitude = []
max_temp = []
humidity = []
cloudiness = []
wind_spd = []
country = []
date = []

units = 'imperial'

print("Beginning Data Retrieval")
print('-------------------------------')

# Initialize a counter for city number
city_count = 0
for city in cities:
    
#     if city_count < 11:
    
    # Set a sleep time of 1 second to ensure the API request limit of 60 per minute is not exceeded
    time.sleep(1)
    # construct the query URL, make the request, and JSON-ify the response
    query_url = f'https://api.openweathermap.org/data/2.5/weather?q={city}&appid={weather_api_key}&units={units}'
    weather_response = requests.get(query_url)
    weather_json = weather_response.json()
    # Open Weather does not necessarily have data for all cities. If a 404 error is encountered, report the error and move to the next city. Otherwise, proceed to print the processing information and extract the relevant data from the JSON response. Append the datapoints to the lists created above
    if weather_json['cod'] == '404':
        print(f"No data available for {city}. Trying next city.")
    else:

        print(f"Processing Record {city_count + 1}: {city}")
        cities_added.append(city)
        city_ID.append(city_count)
        latitude.append(weather_json['coord']['lat'])
        longitude.append(weather_json['coord']['lon'])
        max_temp.append(weather_json['main']['temp_max'])
        humidity.append(weather_json['main']['humidity'])
        cloudiness.append(weather_json['clouds']['all'])
        wind_spd.append(weather_json['wind']['speed'])
        country.append(weather_json['sys']['country'])
        date.append(weather_json['dt'])
        city_count += 1
#     print(weather_json)

#     print(city)
#     print(latitude)
#     print(longitude)
#     print(max_temp)
#     print(humidity)
#     print(cloudiness)
#     print(wind_spd)
#     print(country)
#     print(date)
#     break
print('-------------------------------')
print('Data Retrieval Complete')
print('-------------------------------')
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
# This function takes 7 arguments that allow for the creation of the scatter plot, determination of the linear regression, and plotting of the regression line and the equation
def regress_plot(x_axis, y_axis, title, x_axis_label, y_axis_label, x_coord, y_coord, fig_name):
    slope, y_int, r, pval, stderr = linregress(x_axis, y_axis)
    regress_fit = slope * x_axis + y_int
    plt.scatter(x_axis, y_axis)
    plt.plot(x_axis, regress_fit, '--', color = 'red')
    plt.title(title)
    plt.xlabel(x_axis_label)
    plt.ylabel(y_axis_label)
    line_eq = 'y = ' + str(round(slope, 3)) + 'x + ' + str(round(y_int, 3))
    plt.annotate(line_eq, (x_coord, y_coord), fontsize=15,color="red")
    plt.savefig(fig_name)
    print(f"The r-squared is: {r*r}")
# def annotate_plot(x_coord, y_coord):
# 
```

# Analysis - Latitude vs. Max Temperature
As anticipated, maximum temperatures tend to be warmest for cities near the equator (0 degrees latitude). The regression analysis shows that temperatures then decrease as latitude approaches 90 degrees for the Northern hemisphere or -90 degrees for the Southern hemisphere. Per worldatlas.com, temperatures tend to be warmer near the equator because the sun's energy strikes the equator nearly directly thereby concentrating the warming energy over a smaller area, whereas the sun's rays strike the polar regions at an angle, spreading the energy over a larger area. Additionally, while the equatorial regions of the Earth do no vary in orientation with respect to the sun throughout the seasons, the polar regions spend half of the hear oriented away from the sun, contributing to their receiving indirect sunlight.

# Analysis - Latitude vs. Humidity
Most are familiar with the equatorial regions of the Earth being referred to as "The Tropics", due in part to their year-long warm temperatures and high humidity. The greater amount of solar heat that the tropics receive results in the evaporation of more water from the oceans, and the higher air temperatures increase the saturation point. Interestingly, however, the regression analysis reveals that the correlation between latitude and humidity appears to be quite weak for cities in both the Northern and Southern hemispheres. In fact, both plots suggest that humidity actually increases when moving away from the equator and toward the poles, although again the relationship is very weak. One possible explanation for this trend is that cities that are located inland (where ocean water evaporation does not occur) will tend to be less humid than cities near the coasts. Given that there are almost certainly more inland cities on the globe than coastal cities, our current dataset may simply reflect a larger sampling of inland cities.

# Analysis - Latitude vs. Cloudiness
Similar to humidity, the correlation between latitude and cloudiness appears to be very weak for both hemispheres. Both plots suggest that cloudiness is lowest near the equator and increases slightly toward the polar regions, but the relationship is nowhere near strong enough to suggest a trend. Indeed, clouds will form in any area where water can evaporate at the surface, rise to higher elevations, and condense into water droplets or ice crystals due to the lower air pressure and temperatures. Per NASA, clouds tend to form more frequently in the region 60 degrees north or south of the equator (https://earthobservatory.nasa.gov/images/85843/cloudy-earth#:~:text=Clouds%20also%20tend%20to%20form,and%20south%20of%20the%20equator.&text=Since%20air%20descends%20between%20about,are%20common%20at%20this%20latitude.) However, NASA has the advantage of sampling cloud data from anywhere on the globe, whereas our data is limited to land-based cities. It is therefore not surprising that our plots do not appear to reflect this pattern.
